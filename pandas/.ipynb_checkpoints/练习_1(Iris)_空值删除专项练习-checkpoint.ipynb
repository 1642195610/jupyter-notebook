# Iris(鸢尾花数据集)

```python
import pandas as pd
import numpy as np
```

### 从下述地址加载数据并用变量接收

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
#用pd.read_csv()方法读取数据
iris = pd.read_csv(url)
# 查看数据的前5行
iris.head()
```

### 数据集没有列名,给数据添加列名

```python
#我们读取的是鸢尾花数据集,包含4列特征值和1列标签值即鸢尾花的种类
# 1. sepal_length (in cm) 花瓣长度
# 2. sepal_width (in cm) 花瓣宽度
# 3. petal_length (in cm) 花萼长度
# 4. petal_width (in cm) 花萼宽度
# 5. class 类别

#请按照上述英文或者中文给数据集添加列名
```

```python
# columns = ["sepal_length",
#            "sepal_width",
#            "petal_length",
#            "petal_width",
#            "class"]
# columns
```

```python
columns = ["花瓣长度",
           "花瓣宽度",
           "花萼长度",
           "花萼宽度",
           "类别"]
columns
```

```python
iris.columns = columns
```

```python
iris.columns
```

```python
iris
```

### 判断数据有无空值

```python
pd.isnull(iris.sum())
```

```python
#hint:空值用isnull()方法判断,用sum()方法计算每列空值个数的和,注意axis的用法
```

```python
pd.isnull(iris).sum()
```

###  把花萼长度列的第11行到30行的值设为空

```python
import warnings
warnings.filterwarnings(action='ignore')
iris['花萼长度'][11:31] = np.nan
```

```python
iris.head(20)
```

```python
# 把花萼长度列的第11行到30行的值设为空

# 显示前20行数据
```

### 把空值填充为1

```python
iris.fillna(1,inplace = True)
iris.head(20)
```

```python
# 用fillna()方法填充空值,用shift+tab查看fillna()如何使用
```

```python
iris.dropna(how = "any",inplace = True)
iris.head(5)
```

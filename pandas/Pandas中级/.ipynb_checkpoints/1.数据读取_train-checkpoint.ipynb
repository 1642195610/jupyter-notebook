# 其他数据读取

```python
file = 'text.txt'
with open(file,'r')as f:
    lines = f.read()
    print(lines)
```

```python
with open(file,'r')as f:
    line = f.readlines()
    print(line)
```

```python
with open(file,'r')as f:
    print(f.tell())
    line = f.readline()
    while line:
        print(line)
        print(f.tell())
        line = f.readline()
```

```python
# 封装函数
def connect_c(host,
              user,
              password,
              post,
              database,
              charset,
              sql):
    import pymysql
    import pandas as pd
    ini = {
        'host':host,
        'user':user,
        'password':password,
        'port':post,
        'database':database,
        'charset':charset
             }
    sql = sql
    connect = pymysql.connect(**ini)
    cursor = connect.cursor()
    cursor.execute(sql)
    data = cursor.fetchall()
    df = pd.DataFrame(list(data))
    return df
    cursor.colse()
    connect.colse()
```

```python
# 调用函数
connect_c(
    'localhost',
    'root',
    '7758258',
    3306,
    'db',
    'utf8',
    'select * from 贷款表'
          ).tail()
```

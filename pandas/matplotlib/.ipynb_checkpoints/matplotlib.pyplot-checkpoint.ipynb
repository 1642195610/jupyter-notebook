# 1. 线图

# 数据变化趋势

# .plot 线图

```python
x = np.linspace(0,10,30)
# 生成0 - 10 之间的30个小数
```

```python
x
```

```python
# lw 线宽
# color 颜色: k:黑色
# linestyle 线形
```

```python
plt.plot(
    x, # x轴
    np.sin(x), # y轴
    marker = '8',# 线的标记点
    lw=3, #线宽
    color = 'b', #颜色
    linestyle = 'solid', #线形
    label = 'sin(x)' #图例
)
plt.plot(
    x, # x轴
    np.cos(x), # y轴
    marker = '2',# 线的标记点
    lw=3, #线宽
    color = 'r', #颜色
    linestyle = 'solid', #线形
    label = 'cos(x)' #图例
)
plt.legend() # 显示图例
plt.axis('equal') # 扁平
```

```python

plt.plot(
    x, # x轴
    np.sin(x), # y轴
    '-og', # fmt
    label = 'sin(x)'
)
plt.legend()
```

```python
plt.plot(x, 
         np.sin(x),
         lw=2,color = 'b',
         linestyle = 'solid')
plt.plot(x, 
         np.cos(x),
         lw=2,color = 'b',
         linestyle = '-.')
```

```python
plt.plot(x, 
         np.sin(x),
         lw=2,color = '#FF345F',
         linestyle = 'solid')
```

```python
plt.plot(x, 
         np.sin(x),
         lw=2,color = 'b',
         linestyle = 'solid')
plt.plot(x, 
         np.cos(x),
         lw=2,color = 'b',
         linestyle = '-.')
```

```python
plt.plot(x, 
         np.sin(x),
         lw=2,color = 'b',
         linestyle = 'solid')
plt.plot(x, 
         np.cos(x),
         lw=2,color = 'b',
         linestyle = '-.',
         marker = '2')
```

```python
plt.plot(x, 
         np.cos(x),
         marker = '8',lw=2,color = 'b',
         linestyle = '-.'
)
```

```python
plt.plot(
    x,
    np.sin(x),
    ':^b',
    lw = 3,
    label = 'sin(x)'
)
plt.legend()
plt.axis('equal')
```

# 2.散点图

# 两个变量之间的关系

# .scatter 散点图

```python
x = 2 * np.random.rand(100,1)
```

```python
x
```

```python
y = 4 + 3 * x + np.random.randn(100,1)
```

```python
y
```

```python
plt.scatter(
    x,
    y,
    alpha = 0.3,# 透明度
    s = 1000,# 里面的大小(蓝色)
    c = 'blue',#标记点的颜色(蓝色)
    linewidths = 30,# 膜的大小(红色)
    edgecolors = 'red' # 膜的颜色(红色)
)
```

```python
plt.scatter(
    x,
    y,
    alpha = 0.3,# 透明度
    s = 100,# 里面的大小(蓝色)
    c = 'blue',#标记点的颜色(蓝色)
    linewidths = 10,# 膜的大小(红色)
    edgecolors = 'red' # 膜的颜色(红色)
)
```

```python
plt.scatter(
    x,
    y,
    alpha = 0.3,# 透明度
    s = 100,# 里面的大小(蓝色)
    c = 'blue',#标记点的颜色(蓝色)
    linewidths = 30,# 膜的大小(红色)
    edgecolors = 'red' # 膜的颜色(红色)
)
```

# 4 条形图

# 用来计数

# .bar 条形图

```python
data = [5,20,4,6,8]
plt.bar(
    ['a','b','c','d','e'],# x轴刻度标签
    data # 数据
)
plt.show()
```

```python
data = [5,20,30,12,8]
plt.bar(
    ['a','b','c','d','e'],# x轴刻度标签
    data # 数据
)
plt.show()
```

```python
size = 5
x = np.arange(size)
a = np.random.random(size)
b = np.random.random(size)

plt.bar(x,a,label = 'a')
plt.bar(x,b,bottom = a,label= 'b')
plt.legend()
plt.show()
```

# 5.子图

# .subplot 子图

```python
# 第一种
for i in range(1,5):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize = 15,ha = 'center')
```

```python
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize = 15,ha = 'center')
```

```python
# 第二种 创建画布对象
fig = plt.figure()
# 调整子图距离
fig.subplots_adjust(hspace = 0.4,wspace = 0.4)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5,0.5,str((2,3,i)),fontsize = 15,ha = 'center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace = 0.9,wspace = 0.9)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5,0.5,str((2,3,i)),fontsize = 15,ha = 'center')
```

```python
# 第三种
fig,ax = plt.subplots(3,3,sharex = 'col',sharey = 'row')
```

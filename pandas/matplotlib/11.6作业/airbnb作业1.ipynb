```python
# 把年龄列离散化
div = 15
def get_age(age):
    # 将连续型转换为离散型
    if age < 0:
        return 'NaN' #空值
    elif (age < div):
        return div #如果年龄小于15岁，那么返回15岁
    elif (age <= div * 2):
        return div*2 #如果年龄大于15小于等于30岁，则返回30岁
    elif (age <= div * 3):
        return div * 3
    elif (age <= div * 4):
        return div * 4
    elif (age <= div * 5):
        return div * 5
    elif (age <= 110):
        return div * 6
    else:
        return 'Unphysical' #非正常年龄
data['age'] = data['age'].apply(get_age)
data.age.value_counts()
```

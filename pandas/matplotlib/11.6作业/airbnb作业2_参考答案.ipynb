### 用户行为数据

```python
import pandas as pd
sessions = pd.read_csv('sessions.csv')
sessions.head(10)
```

```python
# 更改列名
sessions.rename(columns={'user_id':'id'},inplace=True)
sessions.head()
```

```python
sessions.isnull().sum()
```

```python
sessions.info(verbose=True,null_counts=True)
```

```python
sessions.dropna(subset=['id'],axis=0,inplace=True)
sessions.head()
```

```python
sessions.shape[0]
```

```python
sessions.isnull().sum()
```

```python
# 缺失值处理
sessions.action = sessions.action.fillna('NAN')
sessions.action_type = sessions.action_type.fillna('NAN')
sessions.action_detail = sessions.action_detail.fillna('NAN')
sessions.isnull().sum()
```

### 重尾分布和对数变换

```python
import matplotlib.pyplot as plt
import numpy as np
# 导入scipy包
from scipy import stats
```

```python
# 查看用户行为的分布
sessions.action.value_counts()
```

```python
sessions.action.value_counts().plot(kind='hist',bins=50)
```

- 行为次数的分布是个长尾分布,长尾分布的特点是均值和方差可能无限大

上图是每月电力生产数

左边是正常数据，可以看到随着时间推进，电力生产也变得方差越来越大，即越来越不稳定

这种情况下常有的分析假设(误差服从独立同分布的正态分布)经常就不会满足

必须让数据尽量满足假设，让方差恒定，即让波动相对稳定

而这种目的可以通过对数转换做到

#### 对数变换的特点
- 用数学语言来说,对数变换可以将[1,10]映射到[0,1],将[10,100]映射到[1,2].
- 用通俗的语言来说,对数变换可以对大数值的范围进行压缩,对小数值的范围进行扩展
- 用高逼格的语言来说,对数变换可以压缩分布高端的长尾,可以扩展分布低端的头部

```python
np.log2(sessions.action.value_counts()).plot(kind='hist',bins=50)
```

```python
np.log10(sessions.action.value_counts()).plot(kind='hist',bins=50)
```

```python
# Box-Cox变换可以明显地改善数据的正态性、对称性
pd.DataFrame(stats.boxcox(sessions.action.value_counts())[0]).plot(kind='hist',bins=50)
```

###  处理各种特征

```python
sessions.action.value_counts()
```

- 从上述结果可以看到不同的行为共有360种

```python
sessions.columns
```

```python
# 第一步，计算各用户每个action的次数
# data.pivot_table()
pt_action_count = sessions.pivot_table(values='device_type',columns='action',index='id',aggfunc='count')
pt_action_count.head()
```

```python
pt_action_count.columns
```

- 透视表的列信息和上面验证的行为种类长度一致

```python
# 第二步计算各用户所有action的均值和方差
pt_action_count['act_mean'] = pt_action_count.mean(axis=1)
pt_action_count['act_std'] = pt_action_count.std(axis=1)
pt_action_count.head()
```

```python
# 第三步计算各个用户action_type的次数
pt_actiontype_count = sessions.pivot_table(values='secs_elapsed',columns='action_type',index='id',aggfunc='count')
pt_actiontype_count.head()
```

```python
import gc
```

```python
gc.collect()
```

```python
 pt_actiontype_count.stack()
```

```python
# 第四步计算各个用户action_type的均值和方差
pt_actiontype_ms = pt_actiontype_count.stack().groupby(level=0).agg([np.mean,np.std])
pt_actiontype_ms.head()
```

```python
# 连接两个表
pt_actiontype_count = pd.concat([pt_actiontype_count,pt_actiontype_ms],axis=1)
pt_actiontype_count.head()
```

```python
# 第五步计算各个用户action_detail的次数
pt_actdetail_count = sessions.pivot_table(values='secs_elapsed',columns='action_detail',index='id',aggfunc='count')
pt_actdetail_count.head()
```

```python
# 第六步计算各个用户action_detail的均值方差
pt_actdetail_ms = pt_actdetail_count.stack().groupby(level=0).agg([np.mean,np.std])
pt_actdetail_ms.head()
```

```python
# 连接表
pt_actdetail_count = pd.concat([pt_actdetail_count,pt_actdetail_ms],axis=1)
pt_actdetail_count.head()
```

```python
# 第七步计算各个用户device_type登录的次数
pt_devtype_count = sessions.pivot_table(values='action',columns='device_type',index='id',aggfunc='count')
pt_devtype_count.head()
```

```python
# 第八步计算各个用户device_type登录次数的均值方差
pt_devtype_ms = pt_devtype_count.stack().groupby(level=0).agg([np.mean,np.std])
pt_devtype_ms.head()
```

```python
# 表连接
pt_devtype_count = pd.concat([pt_devtype_count,pt_devtype_ms],axis=1)
pt_devtype_count.head()
```

```python
# 第九步，按每个用户的均值填充secs_elapsed的空值
sessions['secs_elapsed'] = sessions.groupby('id')['secs_elapsed'].transform(lambda x:x.fillna(x.mean()))
sessions.head()
```

```python
sessions.isnull().sum()
```

```python
sessions[sessions.secs_elapsed.isnull()].head()
```

- 可以观察到有些用户只有一条数据,secs_elapsed同时也为空,填充空值会失效

```python
# 二次填充,0值填充
sessions.secs_elapsed.fillna(0,inplace=True)
sessions.isnull().sum()
```

```python
# 第十步,计算每个用户secs_elapsed的总和,均值和方差
pt_sec = sessions.pivot_table(values='secs_elapsed',index='id',aggfunc=['sum','mean','std'])
pt_sec.head()
```

```python
# 去掉一层列索引
pt_sec = pt_sec.droplevel(level=1,axis=1)
pt_sec.head()
```

```python
# 整合特征
feats = pd.concat([pt_action_count,pt_actiontype_count,pt_actdetail_count,pt_devtype_count],axis=0)
```

# 分组

```python
d.groupby('class')['math'].mean()
```

# 分组排序

```python
d.groupby(by = 'class')['math']. \
apply(lambda x:x.sort_values(ascending = False))
# x 是一个组一个组
```

```python
d.groupby(by = 'class')['math']. \
apply(lambda x:x.sort_values(ascending = True))
# x 是一个组一个组
```

# 每组取前三

```python
d.groupby(by = 'class')['math']. \
apply(lambda x:x.sort_values( \
        ascending = False)[:3])
# x 是一个组一个组
```

```python
df = pd.read_excel('sales_transactions.xlsx')
df
```

```python
# 添加一列,求成员组内占比:
#计算每个sku的销量金额对各自订单的占比
# (即exp price占同一个order的比例)
df.groupby('order')['ext price'].sum()
```

```python
df['ext price']
```

```python
df['占比'] = df.groupby('order')['ext price']. \
apply(lambda x:round(x/x.sum()*100,2))
```

```python
df.groupby('order')['ext price']. \
apply(lambda x:round(x/x.sum()*100,2))
```

```python
to = df.groupby('order')['ext price'] \
.sum().rename('sums')
```

### 5. 对每年每月的销售总额和销量总额进行排序

#### 5.1 统计每年每月的销售总额和销售数量总额

```python
s1.info()
```

```python
s1['下单日期'] = pd.to_datetime(s1['下单日期'],format = "%Y-%m-%d")
s1['预计送货日期'] = pd.to_datetime(s1['预计送货日期'],format = "%Y-%m-%d")
s1['实际送货日期'] = pd.to_datetime(s1['实际送货日期'],format = "%Y-%m-%d")
```

```python
s1.info()
```

```python
s1['销售额'] = s1['数量'] * s1['单价']
```

```python
s1.head()
```

```python
s1['销售年份'] = s1['下单日期'].dt.year
s1['销售月份'] = s1['下单日期'].dt.month
```

```python
s1.head()
```

```python
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
s1.columns
```

```python
s1.head()
```

```python
a = s1.groupby(
    by = ['销售年份','销售月份']
)['销售额','数量'].sum()
a
```

#### 5.2 对上面的结果进行降排,先按销售额排销售额相同则按数量排

```python
a.groupby(['销售年份'],group_keys = False).apply(lambda x:x.sort_values(by = ['销售额','数量'],ascending = [False,False]))
```

# 对上述结果取top3

```python
a.groupby(['销售年份'],group_keys = False).apply(lambda x:x.sort_values(by = ['销售额','数量'],ascending = [False,False])[:3])
```

### 7. 查看数据一致性

#### 7.2 从客户名称中提取用户ID(数字)为新的一列,提取省份为新的一列

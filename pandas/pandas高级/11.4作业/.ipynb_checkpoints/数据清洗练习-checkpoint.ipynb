#### 2.3 提取name对应的值,对genres列来说就是提取每个电影的类型(一个电影可能对应多个类型,比如泰坦尼克号可能对应爱情,灾难两个类型)

```python
def names(name):    
    a = data[name].apply(lambda x:json.loads(x))
    data[name] = a
    data[name].values
    b = data[name]
    aa = []
    for i in b:
        if i != []:
            a = []
            for j in range(len(i)):
                a.append(i[j]['name'])
            aa.append(a)
    res = []
    for i in aa:
        res.append("|".join(i))
    ress = pd.DataFrame(res)
    return ress
```

```python
# names('genres')
```

```python
# 定义一个获取name的函数,name之间用竖线隔开
```

```python
def heads(n):
    name = n
    data[name] = names(name)
    return data
```

```python
data = heads('genres')
data.head()
```

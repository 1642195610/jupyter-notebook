```python
from pyecharts.charts import Gauge
from pyecharts import options as opts
```

```python
gauge = (Gauge()
         .add(
             '仪表板',
             [("",30),("",90)],
             is_clock_wise=True
         )
         .set_global_opts(
             title_opts = opts.TitleOpts(title = 'z',subtitle = 'f')
         )
)
gauge.render_notebook()
```

```python
from pyecharts.charts import Pie
```

```python
cate = ['a','b','c','d']
data = [1,3,5,6]
```

```python
pie = (Pie()
       .add(
           '南丁格尔图', #提示框内容
           [z for z in zip(cate,data)],
           radius=['30%','60%'],# 半径
           rosetype='radius' # 扇片大小
       )
       .set_global_opts(
           title_opts = opts.TitleOpts(title = '主标题',subtitle = '副标题')
       )
)
pie.render_notebook()
```

```python
from pyecharts.charts import Grid
from pyecharts.charts import Funnel
from pyecharts.charts import Map
```

```python
cate = ['a','b','c','d','e']
data = [11,23,34,54,56]
```

```python
funnel = (Funnel()
          .add(
              "漏斗图",
              [i for i in zip(cate,data)],
              sort_ = "descending"
          )
          .set_global_opts(
              title_opts = opts.TitleOpts(title = '主标题',subtitle = '副标题')
          )
          
)
funnel.render_notebook()
```

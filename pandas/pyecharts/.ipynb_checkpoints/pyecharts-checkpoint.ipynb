```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
```

# 条状图 Bar

```python
# 条状图 Bar
# 横坐标数据
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu'] # 横坐标数据
data1 = [123, 153, 89, 107, 98, 23]  # 纵坐标数据
data2 = [56, 77, 93, 68, 45, 67] # 纵坐标
```

```python
bar = (Bar()
       .add_xaxis(cate)
       .add_yaxis('电商',data1)
       .add_yaxis('门店',data2)
      )
bar.render_notebook()
```

```python
bar = (Bar()
       .add_xaxis(cate)
       .add_yaxis('电商',data1)
       .add_yaxis('门店',data2)
       .set_global_opts(title_opts=opts.TitleOpts(title='主标题',subtitle='副标题'))
      )
bar.render_notebook()
```

```python
# is_show = False # 图例选项True-->显示,False-->不显示
bar = (Bar()
       .add_xaxis(cate)
       .add_yaxis('电商',data1)
       .add_yaxis('门店',data2)
       .set_global_opts(
           title_opts=opts.TitleOpts(title='主标题',subtitle='副标题'), # 标题选项
            legend_opts=opts.LegendOpts(is_show = False) # 图例选项True-->显示,False-->不显示
                       )
      )
bar.render_notebook()
```

```python
# set_global_opts
# is_show = False # 图例选项True-->显示,False-->不显示
# 工具箱toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical')
# 提示框 tooltip_opts = opts.TooltipOpts(is_show = True, background_color = 'red')
# 缩放框 datazoom_opts = opts.DataZoomOpts(is_show = True)
# 距离左边界的距离 pos_left = '90%'

# set_series_opts
# 数据标签 label_opts=opts.LabelOpts(is_show = False), # 数据值标签每个柱子上的数字
# 数据点 markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ ="max",name='最大值')]), # 数据点
# 数据线 markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ ='min',name = '最小值')]) # 数据线
bar = (Bar()
       .add_xaxis(cate)
       .add_yaxis('电商',data1)
       .add_yaxis('门店',data2)
       .set_series_opts(
           label_opts = opts.LabelOpts(is_show = False), # 数据值标签每个柱子上的数字
           markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ ="max",name='最大值')]), # 数据点
           markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ ='min',name = '最小值')]) # 数据线
       )
       .set_global_opts(
           title_opts = opts.TitleOpts(title='主标题',subtitle='副标题'), # 标题选项
            legend_opts = opts.LegendOpts(is_show = False), # 图例选项True-->显示,False-->不显示
           toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical'), # 工具箱选项
           tooltip_opts = opts.TooltipOpts(is_show = True, background_color = 'red'), # 提示框选项
           datazoom_opts = opts.DataZoomOpts(is_show = True) # 缩放框选项
                       )
      )
bar.render_notebook() # 渲染图像
# bar.render('Bar.html') # 保存成html
```

```python
# set_global_opts
# is_show = False # 图例选项True-->显示,False-->不显示
# 工具箱toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical')
# 提示框 tooltip_opts = opts.TooltipOpts(is_show = True, background_color = 'red')
# 缩放框 datazoom_opts = opts.DataZoomOpts(is_show = True)
# 距离左边界的距离 pos_left = '90%'

# set_series_opts
# 数据标签 label_opts=opts.LabelOpts(is_show = False), # 数据值标签每个柱子上的数字
# 数据点 markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ ="max",name='最大值')]), # 数据点
# 数据线 markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ ='min',name = '最小值')]) # 数据线
bar = (Bar()
       .add_xaxis(cate)# cate: list[str]
       .add_yaxis('电商',data1)# data1: list[int]
       .add_yaxis('门店',data2)# data2: list[int]
       .set_series_opts(
           label_opts = opts.LabelOpts(is_show = False), # 数据值标签每个柱子上的数字
           markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ = 'max',name = '最大值')]) , # 数据点
           markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ ='min',name = '最小值')]) # 数据线
       )
       .set_global_opts(
           title_opts = opts.TitleOpts(title='主标题',subtitle='副标题'), # 标题选项
            legend_opts = opts.LegendOpts(is_show = False), # 图例选项True-->显示,False-->不显示
           toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical'), # 工具箱选项
           tooltip_opts = opts.TooltipOpts(is_show = True, background_color = 'blue'), # 提示框选项
           datazoom_opts = opts.DataZoomOpts(is_show = True), # 缩放框选项
           xaxis_opts = opts.AxisOpts(axislabel_opts = {'rotate':45})# 横坐标旋转45度,以解决横坐标显示不完全的问题
                       )
      )
bar.render_notebook() # 渲染图像
# bar.render('Bar.html') # 保存成html
```

# 折线图 Line

```python
from pyecharts.charts import Line
```

```python
line = (Line()
    .add_xaxis(cate)
    .add_yaxis('图例一',data1,is_smooth = True)# 是否光滑
    .add_yaxis('图例二',data2,is_smooth = True)
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show = False),
        markpoint_opts=opts.MarkPointOpts(data = [opts.MarkPointItem(type_ = 'max',name = '最大值')]),
        markline_opts=opts.MarkLineOpts(data = [opts.MarkLineItem(type_ = 'min',name = '最小值')])
    )
    .set_global_opts(
        title_opts = opts.TitleOpts(title = '主标题',subtitle='副标题'),
        legend_opts = opts.LegendOpts(is_show = True),
        toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical'),
        tooltip_opts = opts.TooltipOpts(is_show = True,background_color = 'blue'),
        datazoom_opts = opts.DataZoomOpts(is_show = True)
    )
)
line.render_notebook()
```

```python
# set_global_opts
# is_show = False # 图例选项True-->显示,False-->不显示
# 工具箱toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical')
# 提示框 tooltip_opts = opts.TooltipOpts(is_show = True, background_color = 'red')
# 缩放框 datazoom_opts = opts.DataZoomOpts(is_show = True)
# 距离左边界的距离 pos_left = '90%'

# set_series_opts
# 数据标签 label_opts=opts.LabelOpts(is_show = False), # 数据值标签每个柱子上的数字
# 数据点 markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ ="max",name='最大值')]), # 数据点
# 数据线 markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ ='min',name = '最小值')]) # 数据线
line = (Line()
       .add_xaxis(cate)# cate: list[str]
       .add_yaxis('电商',data1,is_smooth = True)# 是否光滑 data1: list[int]
       .add_yaxis('门店',data2,is_smooth = True)# 是否光滑 data2: list[int]
       .set_series_opts(
           label_opts = opts.LabelOpts(is_show = False), # 数据值标签每个柱子上的数字
           markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ = 'max',name = '最大值')]) , # 数据点
           markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ ='min',name = '最小值')]) # 数据线
       )
       .set_global_opts(
           title_opts = opts.TitleOpts(title='主标题',subtitle='副标题'), # 标题选项
            legend_opts = opts.LegendOpts(is_show = False), # 图例选项True-->显示,False-->不显示
           toolbox_opts = opts.ToolboxOpts(pos_left = '90%',orient = 'vertical'), # 工具箱选项
           tooltip_opts = opts.TooltipOpts(is_show = True, background_color = 'blue'), # 提示框选项
           datazoom_opts = opts.DataZoomOpts(is_show = True), # 缩放框选项
           xaxis_opts = opts.AxisOpts(axislabel_opts = {'rotate':45})# 横坐标旋转45度,以解决横坐标显示不完全的问题
                       )
      )
line.render_notebook() # 渲染图像
# line.render('Bar.html') # 保存成html
```

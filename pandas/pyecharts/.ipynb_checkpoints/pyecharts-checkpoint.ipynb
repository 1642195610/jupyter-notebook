# 漏斗图 Funnel

```python
cate = ['访问','注册','加入购物车','提交订单','付款成功']
data = [30398,15230,10045,8109,5698]
```

```python
[list(z) for z in zip(cate,data)]
```

```python
[z for z in zip(cate,data)]
```

```python
from pyecharts.charts import Funnel
from pyecharts import options as opts
```

```python
funnel = (Funnel()
          .add(
              '漏斗图',# 提示框内容
              [i for i in zip(cate,data)],# 数据
              sort_ = 'descending',
              label_opts = opts.LabelOpts(position = 'inside')
          )
          .set_global_opts(
              title_opts = opts.TitleOpts(title = '我是主标题',subtitle = '我是副标题'),
              legend_opts = opts.LegendOpts(is_show = True),
              toolbox_opts = opts.ToolboxOpts(pos_left = '96%',orient = 'vartical'),
              tooltip_opts = opts.TooltipOpts(background_color = 'blue')
          )
)
funnel.render_notebook()
```

# 多图组合 Grid

```python
from pyecharts.charts import Grid
from pyecharts.charts import Bar
from pyecharts.charts import Line
from pyecharts import options as opts
import random
```

```python
city = ['北京市','天津市','上海市','重庆市','阿城市','双城市']
city
```

```python
data = [random.randint(50,150) for i in city]
data
```

```python
bar = (Bar()
       .add_xaxis(city)
       .add_yaxis('营业额',data)
       .set_global_opts(
           title_opts = opts.TitleOpts(title = '条状图'),
           xaxis_opts = opts.AxisOpts(axislabel_opts = {'rotate':90})
       )
       .set_series_opts(
           label_opts =  opts.LabelOpts(is_show = True)
       )
)

line = (Line()
       .add_xaxis(city)
       .add_yaxis('营业额',data)
       .set_global_opts(
           title_opts = opts.TitleOpts(title = '折线图'),
           xaxis_opts = opts.AxisOpts(axislabel_opts = {'rotate':90})
       )
       .set_series_opts(
           label_opts =  opts.LabelOpts(is_show = True)
       )
)

grid = (Grid()
        .add(bar,grid_opts = opts.GridOpts(pos_bottom = '60%'))
        .add(line,grid_opts = opts.GridOpts(pos_top = '60%'))
)
grid.render_notebook()
```

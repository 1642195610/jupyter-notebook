```python
import pandas as pd
import numpy as np
```

```python
from pyecharts.charts import Bar
from pyecharts.charts import Line
```

```python
from pyecharts import options as opts
```

```python
data = pd.read_excel('订单明细.xlsx')
data
```

```python
data.dtypes
```

```python
data['下单日期'].dt.month
```

```python
data['月'] = data['下单日期'].dt.month
data
```

```python
# 获取x轴数据标签函数
def x(column: str):    
    d = data.groupby(by = [column],as_index = False)[['金额']].sum()
    data_list_x = list(d[column].apply(lambda x:str(x)))
    return data_list_x

# 获取数据函数
def y(column: str):   
    d = data.groupby(by = [column],as_index = False)[['金额']].sum()
    data_list_data = list(d['金额'].apply(lambda x:x))
    return data_list_data

# 画图函数 线图
def charts_line(column: str):
    from pyecharts.charts import Line
    from pyecharts import options as opts
    line = (Line()
            .add_xaxis(x(column))
            .add_yaxis("每个" + column + "销售金额",y(column),is_smooth=True)
            .set_series_opts(
                label_opts = opts.LabelOpts(is_show = False),
                markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ = "max",name = "最大值")]),
                markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ = "min",name = "最小值")])
            )
            .set_global_opts(
                title_opts = opts.TitleOpts(title = '销售金额',subtitle = "每个" + column),
                legend_opts = opts.LegendOpts(is_show = True),
                toolbox_opts = opts.ToolboxOpts(pos_left = '97%',orient = 'vartical'),
                tooltip_opts = opts.TooltipOpts(is_show = True,background_color = 'blue'),
                datazoom_opts = opts.DataZoomOpts(is_show = True),
                xaxis_opts = opts.AxisOpts(axislabel_opts = {'rotate':45})
            )
    )
    return line.render_notebook()
```

```python
charts_line('月')
```

* 每个月销售金额最多的是6月,最低的是2月

```python
charts_line('销售')
```

* 每个销售代表的销售金额最高的隔壁老王,最低的是晨晨

```python
charts_line('购买产品类别')
```

* 每个购买产品类别销售金额最多的是配件,最低的是服装

```python
charts_line('客户省份')
```

* 每个客户省份销售金额最多的是天津,最低的是辽宁

```python
from pyecharts.charts import Bar
import pandas as pd
import numpy as np
from pyecharts import options as opts
```

```python
x = ['a','b','c']
d = [33,22,44]
```

```python
bar = (Bar()
       .add_xaxis(x)
       .add_yaxis('第一个图例的名字',d)
       .set_series_opts(
           label_opts = opts.LabelOpts(is_show = True),
           markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_ = 'max',name = '最大值')]),
           markline_opts = opts.MarkLineOpts(data = [opts.MarkLineItem(type_ = 'min',name = '最小值')])
       )
       .set_global_opts(
           title_opts = opts.TitleOpts(title = '我是主标题',subtitle = '我是副标题'),
           legend_opts = opts.LegendOpts(is_show = True),
           toolbox_opts = opts.ToolboxOpts(pos_left = '96%',orient = 'vartical'),
           tooltip_opts = opts.TooltipOpts(is_show = True,background_color = 'blue'),
           datazoom_opts = opts.DataZoomOpts(is_show = True),
           xaxis_opts = opts.AxisOpts(axislabel_opts = {'ratote':60})
       )
)
bar.render_notebook()
```

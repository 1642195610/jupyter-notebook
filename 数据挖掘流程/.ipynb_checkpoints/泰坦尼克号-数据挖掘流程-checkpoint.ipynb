```python
tr.groupby('Initial')['Age'].mean()
```

```python
a = list(tr.groupby('Initial')['Age'].mean().index)
a
```

```python
b = [np.ceil(i) for i in tr.groupby('Initial')['Age'].mean().values]
b
```

```python
for i in range(len(a)):
    tr.loc[(tr['Age'].isnull())&(tr['Initial'] == a[i]),'Age'] = b[i] 
```

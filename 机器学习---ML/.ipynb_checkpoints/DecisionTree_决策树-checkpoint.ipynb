```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import classification_report
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target

print(iris.data)
print(iris.target)
```

```python
x.shape
```

```python
x
```

```python
y
```

```python
clf = DecisionTreeClassifier(criterion='gini',
                            splitter='best',
                            max_depth=4,
                            min_samples_split=5,
                            min_samples_leaf=5
                           )
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.05,random_state = 0)
```

```python
clf.fit(x_tr,y_tr)
```

```python
y_predict = clf.predict(x_te)
y_predict
```

```python
y_te
```

```python
print(f"训练集_SCORE :  {clf.score(x_tr,y_tr)}")
print(f"测试集_SCORE :  {clf.score(x_te,y_te)}")
```

```python
print(classification_report(y_te,y_predict))
```

# 找最优超参数

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridsearchcv = GridSearchCV(
    estimator=clf, # 学习器的名字
    param_grid={'max_depth':[3,4,5,6,7,8,9],
                'min_samples_split':[2,3],
                'min_samples_leaf':[3,4,5,6,7,8]
                }, # 最优超参数字典
    cv = 10, # 几折交叉验证
    refit = True, # 
    verbose=2 # 日志等级_详细程度
)
```

```python
gridsearchcv.fit(x,y)
```

```python
pd.DataFrame(gridsearchcv.cv_results_)
```

```python
gridsearchcv.best_estimator_ # 最好的学习器
```

```python
gridsearchcv.best_params_ # 最好的超参数字典
```

```python
gridsearchcv.best_score_ # 最好的成绩
```

```python
gridsearchcv.best_index_ # 最好的
```

```python
from IPython.display import Image
from sklearn import tree
import pydotplus
```

```python
dot_data = tree.export_graphviz(
    gridsearchcv.best_estimator_,
    out_file = None,
    feature_names = ['sepal length', 
                     'sepal width', 
                     'Petal length', 
                     'Petal width'],
    class_names = iris.target_names,
    filled = True,
    rounded = True,
    special_characters = True
                              )
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

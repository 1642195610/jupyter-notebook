```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import os
```

```python
np.random.seed(42)
m = 100
x = 6*np.random.rand(m,1)-3
y = 0.5*x**2+x+2+np.random.randn(m,1)
x_new = np.linspace(-3,3,100).reshape(100,1)
```

```python
root_dir = '.'
model = "model"
def save_png(png_id,is_save = True):
    path = os.path.join(root_dir,'png',model)
    if not os.path.exists(path):
        os.makedirs(path)
    path = os.path.join(path,png_id)
    print(path)
    print('保存中',png_id)
    plt.savefig(path,format = 'png',dpi = 300)    
```

```python
for style,width,degree in (('g-',1,300),('b--',2,2),('r-+',2,1)):
    polybig_features = PolynomialFeatures(degree = degree,include_bias = False)
    std_scaler = StandardScaler()
    mm_scaler = MinMaxScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
        ('poly_features',polybig_features),
        ('std_scaler',std_scaler),
        ('lin_reg',lin_reg),
    ])
    polynomial_regression.fit(x,y)
    y_newbig = polynomial_regression.predict(x_new)
    plt.plot(x_new,y_newbig,style,label = str(degree),linewidth = width)
    
plt.plot(x,y,'b.',linewidth=3)
plt.legend(loc = 'upper left')
plt.xlabel('$x_1$',fontsize = 18)
plt.ylabel("$y$",fontsize = 18)
plt.axis([-3,3,0,10])
save_png('拟合图.png')
plt.show()
```

```python
def plot_learning_curves(model,x,y,c,r):
    x_tr,x_val,y_tr,y_val = train_test_split(x,y,test_size = 0.2,random_state = 10)
    tr_error = []
    val_error = []
    for m in range(1,len(x_tr)):
        model.fit(x_tr[:m],y_tr[:m])
        y_tr_pre = model.predict(x_tr[:m])
        y_val_pre = model.predict(x_val)
        tr_error.append(mean_squared_error(y_tr[:m],y_tr_pre))
        val_error.append(mean_squared_error(y_val,y_val_pre))
        
    plt.plot(np.sqrt(tr_error),c + '--',linewidth = 2,label = 'train_' + str(r))
    plt.plot(np.sqrt(val_error),c + '-',linewidth = 3,label = 'val_' + str(r))
    plt.legend(loc = 'upper right',fontsize = 14)
    plt.xlabel("Training set size",fontsize = 14)
    plt.ylabel('RMSE',fontsize = 14)
```

```python
l = ['r','g','b']
r = [300,2,1]
for i in range(1,4):
    poly_features = PolynomialFeatures(degree = r[i-1],include_bias = False)
    x_poly = poly_features.fit_transform(x)
    
    lin_reg = LinearRegression()
    plot_learning_curves(lin_reg,x_poly,y,l[i-1],r[i-1])
    plt.axis([0,80,0,3])
    save_png('学习曲线图.png')
plt.show()
```

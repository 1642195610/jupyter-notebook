```python
import torch
import numpy as np
from torch import nn
```

```python
class LinearNet(nn.Module):
    def __init__(self,n_feature):
        super(LinearNet,self).__init__()
        self.linear = nn.Linear(n_feature,1)
        for item in self.linear.parameters():
            nn.init.normal_(item,mean = 0,std = 0.01)
    def forward(self,x):
        y = self.linear(x)
        return y
    
linearnet = LinearNet(2)
print(linearnet)
```

```python
for param in linearnet.parameters():
    print(param)
```

```python
net = nn.Linear(2,1)
nn.init.normal_(net.weight,mean=0,std=0.01)
```

```python
nn.init.constant_(net.bias,val = 0)
```

```python
loss = nn.MSELoss()
```

```python
linearnet.parameters()
```

```python
import torch.optim as optim
```

```python
optimizer = optim.SGD(linearnet.parameters(),lr = 0.03)
print(optimizer)
```

```python
import torch.utils.data as Data
```

```python
batch_size = 32
```

```python
features = torch.randn(1000,2,dtype = torch.float32)
true_w = [3,-4.5]
true_b = 4
labels = true_w[0] * features[:,0] + true_w[1] * features[:,1] + true_b
labels += torch.tensor(np.random.normal(0,0.01,size = labels.size()),dtype = torch.float32)
```

```python
dataset = Data.TensorDataset(features,labels)
```

```python
data_iter = Data.DataLoader(dataset,batch_size,shuffle = True)
```

```python
epochs = 20
for epoch in range(1,epochs + 1):
    for x,y in data_iter:
        # 1.预测
        pred = linearnet(x)
        # 2.损失
        l = loss(pred,y.view(-1,1))
        # 3.梯度清零
        optimizer.zero_grad()
        # 4.反向传播
        l.backward()
        # 5.更新网络参数
        optimizer.step()
    print(f"epoch{epoch},loss:   {l.item()}")
```

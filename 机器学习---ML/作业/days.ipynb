```python
# 4.
def age_cut(col):
    if col >= 20 and col < 40:
        return "年轻"
    elif col < 60:
        return "中年"
    elif col < 80:
        return "中老年"
    else:
        return "老年"
```

```python
# 定义年龄分箱函数
```

```python
# 5.
patients['年龄段'] = patients['年龄'].apply(lambda x:age_cut(x))
patients.drop(labels = ['年龄'],axis = 1,inplace = True)
patients.head()
```

```python
# 10.
col = ['性别','年龄段','胖瘦','家庭规模','医疗资源','发病时期']
col

def drbm(data,col: str):
    prefix = col + "_"
    df = pd.get_dummies(data[col],prefix = prefix)
    da = pd.concat((data,df),axis = 1,ignore_index = False)
    return da

for i in col:
    patients = drbm(patients,i)
    patients.drop(labels = i,axis = 1,inplace = True)
patients.head()
```

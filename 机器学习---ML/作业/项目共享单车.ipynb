```python
# 20.
def cut_s(col):
    df[col] = pd.cut(df[col],bins = 4,labels = ['大','中','小','微小'])
    return df

col = ['temp','atemp','humidity','windspeed']
for i in col:
    cut_s(i)
df
```

```python
# 21.
col = ['season','weather','week','temp','atemp','humidity','windspeed','hour','month']
col

def drbm(data,col: str):
    prefix = col + "_"
    df = pd.get_dummies(data[col],prefix = prefix)
    da = pd.concat((data,df),axis = 1,ignore_index = False)
    return da

for i in col:
    df = drbm(df,i)
    df.drop(labels = i,axis = 1,inplace = True)
df.head()
```

```python
# 22.
col = [x for x in df.columns if x not in ['count','hour','year_month','date','month']]
col
```

```python
# 23.
x = df.loc[:,col]
y = df.loc[:,'count']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
from sklearn.linear_model import Ridge
from sklearn import metrics
from math import sqrt
```

```python
# 24.
ridge = Ridge(alpha=1.0)
ridge.fit(x_train,y_train)

y_predict_train = ridge.predict(x_train)
y_train_rmse = sqrt(metrics.mean_squared_error(y_train,y_predict_train))
y_train_score = ridge.score(x_train,y_train)

y_predict_test = ridge.predict(x_test)
y_test_rmse = sqrt(metrics.mean_squared_error(y_test,y_predict_test))
y_test_score = ridge.score(x_test,y_test)

print('训练集RMSE: {0}, 评分: {1}'.format(y_train_rmse, y_train_score))
print('测试集RMSE: {0}, 评分: {1}'.format(y_test_rmse, y_test_score))
```

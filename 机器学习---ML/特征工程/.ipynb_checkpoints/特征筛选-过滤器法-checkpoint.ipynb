```python
from sklearn.model_selection import GridSearchCV
```

```python
def getb(model, params,x,y):
    grid = GridSearchCV(model,
                        param_grid = params,
                        cv = 5,
                        refit = True)
    grid.fit(x,y)
    print(f"最好的得分: {grid.best_score_}")
    print(f"最好的最优超参数: {grid.best_params_}")
```

```python
from sklearn.base import TransformerMixin, \
BaseEstimator
```

```python
# res 标签列
# lis 数据 > thr的列
# thr 阈值
class CustomCorrlationChooser(TransformerMixin,BaseEstimator):
    def __init__(self,res, lis = [],thr = None):
        self.res = res
        self.lis = lis
        self.thr = thr
        
    def fit(self,x,*_):
        df = pd.concat([x,self.res],axis = 1)
        t = df.columns[df.corr()[df.columns[-1]].abs() > self.thr]
        self.lis = [i for i in t if i in x.columns]
        return self
    
    def transform(self,x):
        return x[self.lis]
    
    def transform_fit(self,x):
        df = pd.concat([x,self.res],axis = 1)
        t = df.columns[df.corr()[df.columns[-1]].abs() > self.thr]
        self.lis = [i for i in t if i in x.columns]
        return x[self.lis]
```

```python
ccc = CustomCorrlationChooser(res = ys,thr= .2)
ccc.fit(xs)
ccc.lis
```

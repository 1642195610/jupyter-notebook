# 有监督分箱

### 基于信息熵的分箱

```python
import entropy_based_binning as ebb # 基于信息熵的一种有监督分箱方法
```

```python
y = y.reshape(150,-1)
```

```python
df = np.concatenate((x*10,y),axis = 1)
```

```python
df
```

```python
df1 = ebb.bin_array(df,nbins=3,axis = 0)
```

```python
df1
```

```python
xd4 = df1[:,:2]
sns.scatterplot(xd4[:,0],xd4[:,1])
```

```python
clf = DecisionTreeClassifier()
s4 = cross_val_score(clf,xd4,y,cv = 5)
```

```python
print(np.mean(s4))
```

```python
df2 = ebb.bin_array(df,nbins=2,axis = 0)
```

```python
xd5 = df2[:,:2]
sns.scatterplot(xd5[:,0],xd5[:,1])
```

```python
s5 = cross_val_score(clf,xd5,y,cv = 5)
print(np.mean(s5))
```

```python
from mdlp.mdlp import MDLP
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
mdlp = MDLP(categorical_features = np.arange(X.shape[1]),
            min_depth=5,
            random_state=0,
            shuffle=False)
conv_X = mdlp.fit_transform(X, y)
score = cross_val_score(clf,conv_X,y,cv = 5)
print(np.mean(score))
```

```python
import numpy as np
```

```python
# 风险函数
def f(x):
    return x**2-4*x+4
```

```python
# 损失函数
def h(x):
    return 2*x-4
```

```python
theta = 16
setp = 0.1
count = 0
error = 1e-5
```

```python
while True:
    s = setp * h(theta)
    if s < error:
        break
    else:
        theta = theta - s
        count += 1
    print("梯度下降迭代第{}次,原始值为{},更新后的值为{}".format(count,theta,h(theta)))
print("--------------------------------")
print(f"迭代次数为{count}")
print(f"极值点为{f(theta)}")
```
